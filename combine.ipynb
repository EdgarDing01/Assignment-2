````
import pandas as pd

# Load both datasets
try:
    df_projects = pd.read_csv("cleaned_world_bank.csv")
    df_macro = pd.read_csv("world_bank_macroeconomic_panel_1990_2025.csv")

    # --- Pre-processing ---

    # Keep track of original project count
    original_project_count = len(df_projects)

    # 1. Handle missing values in key columns
    # Drop rows from projects where 'approval_fy' or 'country_/_economy' is null
    df_projects = df_projects.dropna(subset=['approval_fy', 'country_/_economy'])

    # 2. Convert 'approval_fy' to integer to match 'Year' data type
    df_projects['approval_fy'] = df_projects['approval_fy'].astype(int)

    # Projects remaining after cleaning keys
    projects_to_merge_count = len(df_projects)

    # --- Merging ---

    # Perform a left merge
    # This keeps all rows from df_projects
    # and adds data from df_macro where the keys match.
    merged_df = pd.merge(
        left=df_projects,
        right=df_macro,
        left_on=['country_/_economy', 'approval_fy'],
        right_on=['CountryName', 'Year'],
        how='left'
    )

    # --- Post-Merge Inspection & Saving ---

    print("--- Merged DataFrame Info ---")
    merged_df.info()

    print("\n--- Merged DataFrame Head ---")
    print(merged_df.head())

    # Check merge success
    # We check for a non-null column from the right table (e.g., 'GDP_USD' or 'CountryName')
    successful_matches = merged_df['CountryName'].notna().sum()

    print("\n--- Merge Summary ---")
    print(f"Original projects: {original_project_count}")
    print(f"Projects with valid keys (country & approval year): {projects_to_merge_count}")
    print(f"Projects successfully matched with macro data: {successful_matches}")
    print(f"Projects without a macro data match: {projects_to_merge_count - successful_matches}")

    # Save the merged DataFrame to a new CSV file
    output_filename = "combined_world_bank_data.csv"
    merged_df.to_csv(output_filename, index=False)

    print(f"\nSuccessfully combined datasets and saved to '{output_filename}'")

except FileNotFoundError as e:
    print(f"Error loading file: {e}")
except Exception as e:
    print(f"An error occurred during merging: {e}")
````
